```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target  # Add target column as species
data['species'] = data['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})  # Map target to species names

# Display the first few rows of the dataset
print("First 5 Rows of the Dataset:\n", data.head())

# Display dataset information
print("\nDataset Information:\n")
data.info()

# Check for missing values
print("\nMissing Values:\n", data.isnull().sum())
```

```python
# Step 1: Compute basic statistics
print("\nBasic Statistics of Numerical Columns:\n")
print(data.describe())

# Step 2: Group by 'species' and compute the mean for each species
print("\nMean Measurements Grouped by Species:\n")
grouped_data = data.groupby('species').mean()
print(grouped_data)

# Step 3: Identify interesting patterns
print("\nInsights:\n")
print("Setosa seems to have the smallest petal dimensions, while Virginica has the largest.")
```

```python
import matplotlib.pyplot as plt

# Plot line chart for sepal length trends
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['sepal length (cm)'], label="Sepal Length", color='blue')
plt.title("Trend of Sepal Length Across Observations", fontsize=16)
plt.xlabel("Index", fontsize=12)
plt.ylabel("Sepal Length (cm)", fontsize=12)
plt.legend()
plt.grid()
plt.show()
```

```python
# Plot bar chart for average petal length by species
grouped_data['petal length (cm)'].plot(kind='bar', color=['orange', 'green', 'purple'], alpha=0.8)
plt.title("Average Petal Length by Species", fontsize=16)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Petal Length (cm)", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
import seaborn as sns

# Plot histogram of petal width
plt.figure(figsize=(8, 6))
sns.histplot(data['petal width (cm)'], bins=10, kde=True, color='teal', edgecolor='black')
plt.title("Distribution of Petal Width", fontsize=16)
plt.xlabel("Petal Width (cm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y')
plt.show()
```

```python
# Scatter plot for sepal length vs petal length by species
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=data['sepal length (cm)'],
    y=data['petal length (cm)'],
    hue=data['species'],
    palette='viridis',
    s=100
)
plt.title("Sepal Length vs Petal Length", fontsize=16)
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Petal Length (cm)", fontsize=12)
plt.legend(title='Species')
plt.grid()
plt.show()
```

```python
# Iris Data Analysis Assignment
### Author: Sharon Kambe

This notebook explores the Iris dataset to understand the relationships between sepal and petal measurements across three Iris species: Setosa, Versicolor, and Virginica. The analysis includes:
- Data Loading and Exploration
- Basic Statistical Analysis
- Visualizations
- Insights and Key Findings
```

```python
# Import necessary libraries
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target  # Add target column as species
data['species'] = data['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})  # Map target to species names

# Display the first few rows of the dataset
print("First 5 Rows of the Dataset:\n", data.head())

# Display dataset information
print("\nDataset Information:\n")
data.info()

# Check for missing values
print("\nMissing Values:\n", data.isnull().sum())
```

```python
# Basic statistics
print("\nBasic Statistics of Numerical Columns:\n")
print(data.describe())

# Grouped means by species
print("\nMean Measurements Grouped by Species:\n")
grouped_data = data.groupby('species').mean()
print(grouped_data)
```

```python
### Insights from Basic Analysis:
- Setosa has the smallest petal dimensions, while Virginica has the largest.
- Sepal width is widest for Setosa and narrows for Versicolor and Virginica.
- Clear distinctions in measurements between species indicate well-defined clusters.
```

```python
import matplotlib.pyplot as plt

# Line Chart
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['sepal length (cm)'], label="Sepal Length", color='blue')
plt.title("Trend of Sepal Length Across Observations", fontsize=16)
plt.xlabel("Index", fontsize=12)
plt.ylabel("Sepal Length (cm)", fontsize=12)
plt.legend()
plt.grid()
plt.show()
```

```python
# Bar Chart
grouped_data['petal length (cm)'].plot(kind='bar', color=['orange', 'green', 'purple'], alpha=0.8)
plt.title("Average Petal Length by Species", fontsize=16)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Petal Length (cm)", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
import seaborn as sns

# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data['petal width (cm)'], bins=10, kde=True, color='teal', edgecolor='black')
plt.title("Distribution of Petal Width", fontsize=16)
plt.xlabel("Petal Width (cm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=data['sepal length (cm)'],
    y=data['petal length (cm)'],
    hue=data['species'],
    palette='viridis',
    s=100
)
plt.title("Sepal Length vs Petal Length", fontsize=16)
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Petal Length (cm)", fontsize=12)
plt.legend(title='Species')
plt.grid()
plt.show()
```

```python
### Key Findings:
1. **Setosa** has distinctively small petals and wider sepals compared to the other species.
2. **Virginica** has the largest petal dimensions and longest sepals.
3. **Versicolor** lies between Setosa and Virginica, with intermediate measurements.
4. The scatter plot shows clear clustering based on species, indicating separable patterns in the dataset.

### Conclusion:
The Iris dataset demonstrates well-defined separations between the three species. This analysis provides insights into the relationships between sepal and petal measurements, highlighting patterns useful for classification tasks.
```
